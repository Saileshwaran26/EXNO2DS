```python
dt=pd.read_csv("drive/MyDrive/SEM 1 /19AI403 INTRODUCTION TO DATA SCIENCE/COLAB/titanic_dataset.csv")

```

```python
dt.info()
print(f"Number of rows: {dt.shape[0]}")
print(f"Number of columns: {dt.shape[1]}")
```

```python
dt.set_index('PassengerId', inplace=True)
```

```python
print(dt['Sex'].value_counts())
print(dt['Embarked'].value_counts())
```

```python
sns.countplot(x='Survived', data=dt)
plt.title('Survival Count')
plt.show()
```

```python
print(dt['Pclass'].unique())
```

```python
dt.rename(columns={'Sex': 'Gender'}, inplace=True)
print(dt.head())
```

```python
sns.catplot(x="Pclass", hue="Survived", kind="count", data=dt)
plt.title('Pclass vs Survived')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=dt)
plt.title('Age Distribution by Survival')
plt.show()
```

```python
sns.boxplot(x="Pclass", y="Age", hue="Gender", data=dt)
plt.title('Pclass vs Age by Gender')
plt.show()
```

```python
sns.catplot(x="Pclass", hue="Survived", col="Gender", kind="count", data=dt)
plt.title('Pclass vs Survived by Gender')
plt.show()
```

```python
numerical_features = dt.select_dtypes(include=['number'])

plt.figure(figsize=(10, 8))
sns.heatmap(numerical_features.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Heatmap of Correlations')
plt.show()
```

```python
sns.pairplot(dt, hue="Survived", palette="coolwarm")
plt.show()
```
